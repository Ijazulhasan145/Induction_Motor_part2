## Logistic Regression (Binary classification)

```python
# IMPORT the necessary libraries
import  matplotlib.pyplot as plt
from sklearn.preprocessing import LabelEncoder
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
```

```python
# Encode the status label
le = LabelEncoder()
motor_dataset['Status'] = le.fit_transform(motor_dataset['Status'])
motor_dataset
```

```python
# split the data into features and label
features = motor_dataset.drop('Status', axis=1)
label = motor_dataset['Status']
```

```python
# Now Split the data into train and test
x_train, x_test, y_train, y_test = train_test_split(features, label, test_size = 0.2)
```

```python
# now make sigmoid function for binary class
def sigmoid(z):
  return 1 / (1 + np.exp(-z))
```

```python
# Logistic Regression function
def logistic_regression(features, label, learning_rate, num_iterations):
    # Ensure `label` is a column vector (shape: (num_samples, 1))
    label = label.reshape(-1, 1)

    # Initialize parameters
    num_samples, num_features = features.shape
    weights = np.zeros((num_features, 1))  # Shape: (num_features, 1)

    # Gradient Descent
    for i in range(num_iterations):
        # Compute the linear model
        z = np.dot(features, weights)  # Shape: (num_samples, 1)

        # Compute the activation (sigmoid function)
        predictions = sigmoid(z)  # Shape: (num_samples, 1)

        # Compute the cost
        cost = -np.mean(label * np.log(predictions) + (1 - label) * np.log(1 - predictions))

        # Compute the gradient
        gradient = np.dot(features.T, (predictions - label)) / num_samples  # Shape: (num_features, 1)

        # Update the weights
        weights -= learning_rate * gradient

        # Print the cost every 100 iterations
        if i % 100 == 0:
            print(f"Iteration {i}: Cost = {cost:.4f}")

    return weights
```

```python
# now make the function for prediction
def predict(features, weights):
  m = features.shape[0]
  predictions = np.zeros((m,1))
  weights = weights.reshape(features.shape[1],1)

  # compute the prob of the positive class
  z = np.dot(features, weights)
  predictions = sigmoid(z)

  # Assigning the label using threshold which is 0.5
  predictions[predictions >= 0.5] = 1
  predictions[predictions < 0.5] = 0

  return predictions
```

```python
weights = logistic_regression(x_train.values, y_train.values, learning_rate=0.01, num_iterations=1000)
```

```python
weights
```

```python
plt.plot(range(1000), weights)
plt.xlabel("Iteration")
plt.ylabel("Cost")
plt.title("Gradient Descent")
plt.show()
```

```python
predict_y = predict(x_test.values, weights)
predict_y
```

```python
print("Accuracy:", accuracy_score(y_test, predict_y))
```

```python
cp_report = classification_report(y_test, predict_y)
print(cp_report)
```

```python
predict_y = predict_y.ravel()  # Convert to 1D
y_test = y_test.ravel()        # Convert to 1D

TP = np.sum((predict_y == 1) & (y_test == 1))
FP = np.sum((predict_y == 1) & (y_test == 0))
TN = np.sum((predict_y == 0) & (y_test == 0))
FN = np.sum((predict_y == 0) & (y_test == 1))
sensitivity = TP / (TP + FN)
specificity = TN / (TN + FP)
print("Sensitivity:", sensitivity)
print("Specificity:", specificity)
```

```python
# Now compare it with built-in model
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
model.fit(x_train, y_train)
print("Accuracy:", model.score(x_test, y_test))
```
